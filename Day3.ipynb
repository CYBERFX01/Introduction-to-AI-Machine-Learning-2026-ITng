```python
import pandas as pd
import numpy as np
```

```python
# 1. LOAD DATA
df = pd.read_csv("dataset/dirty_cafe_sales.csv")
```

```python
print("Original Data:")
print(df.head())
```

```python
print("\nDataset Info:")
print(df.info())
```

```python
# 2. STANDARDIZE MISSING VALUES
# Replace common dirty values with NaN
df.replace(["UNKNOWN", "ERROR"], np.nan, inplace=True)
```

```python
print(df.head())
```

```python
print("\nAfter replacing dirty values:")
print(df.isnull().sum())
```

```python
# 3. HANDLE MISSING VALUES
# Separate numeric and categorical columns
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
categorical_cols = df.select_dtypes(include=["object"]).columns
```

```python
# Fill numeric with mean
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())
```

```python
# Fill categorical with mode
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
print("\nAfter handling missing values:")
print(df.isnull().sum())
```

```python
# 4. REMOVE DUPLICATES
before = len(df)
df = df.drop_duplicates()
after = len(df)

print(f"\nduplicates: {before - after}")
```

```python
# 5. FIX DATA TYPES
# Try converting numeric columns properly
for col in df.columns:
    try:
        df[col] = pd.to_numeric(df[col])
    except:
        pass
```

```python
# Convert date column
for col in df.columns:
    if "date" in col.lower():
        df[col] = pd.to_datetime(df[col], errors="coerce")
```

```python
print(df.dtypes)
```

```python
# 6. FINAL CHECK
print("\nFinal Dataset:")
print(df.head())
```

```python
# 7. SAVE CLEANed DATA
df.to_csv("dataset/NEWcleaned_cafe_sales.csv", index=False)
```
